# The model training program using Linear Regression

```python
import turicreate as tc
phones = tc.SFrame('phone_dataset .csv')
def india_price(price):
    india = price * 86.04
    return india

phones['approx_price_INR'] = phones['approx_price_EUR'].apply(india_price)
del phones['approx_price_EUR']
phones = phones.fillna('approx_price_INR', 0)
phones = phones.fillna('weight_g', 0)
phones = phones.fillna('weight_oz', 0)
train, test = phones.random_split(0.85, seed= 0)
model = tc.linear_regression.create(train, target = 'approx_price_INR', features= None, validation_set= test)
```
